## SPLITTING DATE OF JOURNEY INTO DAY MONTH AND YEAR IN INTEGER FORMAT

```python
day = []
month = []
year = []

for date in data['Date_of_Journey']:
    date = str(date).split('/')
    day.append(int(date[0]))
    month.append(int(date[1]))
    year.append(int(date[2]))

data.insert(loc=1, column='JR_Day', value=day)
data.insert(loc=2, column='JR_Month', value=month)
data.insert(loc=3, column='JR_Year', value=year)
```

```python
data.head()
```

```python
data.drop(columns=['Date_of_Journey'], axis=1,inplace=True)
```

```python
data.head()
```

## SPLITTING THE ARRIVAL TIME AND REMOVING THE UNECESSARY DATE 

```python
at = []
for time in data['Arrival_Time']:
    time = str(time).split(' ')[0]
    at.append(time)

```

```python
data.drop(columns=['Arrival_Time'],axis=1, inplace=True)
```

```python
data.insert(loc=8,column='Arrival_Time', value=at)
```

```python
data.head()
```

## NOW SPLITTING ARRIVAL AND DEPARTURE TIME INTO HOURS AND MINUTES

```python
arr_hrs = data['Arrival_Time'].str.split(':').str[0].astype(int)
arr_mins = data['Arrival_Time'].str.split(':').str[1].astype(int)
dep_hrs = data['Dep_Time'].str.split(':').str[0].astype(int)
dep_mins = data['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
data.drop(columns=['Arrival_Time','Dep_Time'], axis=1, inplace=True)
```

```python
data.insert(loc=7,column='Dep_Hours', value=dep_hrs)
data.insert(loc=8,column='Dep_Mins', value=dep_mins)
data.insert(loc=9,column='Arrival_Hours', value=arr_hrs)
data.insert(loc=10,column='Arrival_Min', value=arr_mins)
```

```python
data.info()
```

## CONVERTING THE DATA TYPE OF TOTAL STOPS FROM OBJECT TO INTEGER

```python
data['Total_Stops'] = data['Total_Stops'].map(
    {
        '1 stop'  : 1,
        'non-stop'  : 0,
        '2 stops'  : 2,
        '3 stops'  : 3,
        '4 stops'  : 4
    }
)
```

## CONVERTING THE DURATION INTO MINS

```python
dur = data['Duration']
min_dur = []
for d in dur:
    d = str(d.replace('h','').replace('m', ''))
    d = d.split(' ')

    if len(d) == 1:
        d = int(d[0])
        if d < 60 :
            pass
        else:
            d = d*60
    else:
        d = int(d[0])*60 + int(d[1])
    min_dur.append(d)
```

```python
data.head()
```

```python
data.insert(loc=11, column='Duration in Minutes', value=min_dur)
```

```python
data.head()
```

```python
data.drop(columns=['Duration'],axis=1,inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

rfr = RandomForestRegressor()
X = train_data.iloc[:,:-1]
Y = train_data.iloc[:,-1]
X_train , X_test, y_train, y_test = train_test_split(X,Y,test_size=0.15, random_state=10)
rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test)
print("MSE:", round(mean_squared_error(y_pred,y_test),3))
print("R2:", round(r2_score(y_pred, y_test),2))
print("--------------")


for i in y_pred:
    print(i)

```
