## THIS IS THE ZOMATO DATA SET EDA 

```python
# IMPORTING THE REQUIRED MODULES
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# IMPORTING THE DATA SET 'zomato.csv' and 'country-code.xlsx'
df = pd.read_csv('zomato.csv', encoding='latin-1')
countries = pd.read_excel('Country-Code.xlsx')
```

```python
df
```

```python
countries
```

# Observation
Here we can see the df data frame tells us about the data related to listed/registered restaurants on Zomato from all over the world
Also another data frame - countries tells us the country codes which are mentioned in the df data frame regarding zomato data set

```python
# Now We merge both the data sets into one data set so as to work with both the data sets at one
```

```python
data = pd.merge(df,countries,on='Country Code',how='left')
```

```python
#Merged Data Set
data
```

```python
# Now let us explore all the different columns provided to us in the data set
data.info()
```

```python
# To check the data-types of the columns
data.dtypes
```

```python
# Here we have floating point numbers , integers and objects present in the data set
# Now let us look at if null values are present in the data set
df.isnull().sum()
```

```python
# Cuisines have 9 NULL values present in the data set observations given to us
# to extract only Cuisines:
for i in data.columns:
    if data[i].isnull().sum() > 0:
        print(i, data[i].isnull().sum())
```

```python
#Plotting HeatMap to show the outliers in the Cuisines:
sns.heatmap(data=data.isnull(),yticklabels=False)
```

```python
# Now let us look at the numerical data set provided to us
data.describe()
```

```python
#FROM HERE WE CAN OBSERVATIONS REGARDING AVERAGE COST FOR TWO , PRICE RANGE, AGGREGATE RATING AND VOTES ON THE FIELDS LIKE MAXIMUM / MINIMUM , STD , MEAN AND OTHER QUARTILE VALUES
print('Average Cost for two	:', round(data.describe().loc['mean','Average Cost for two'],2))
```

```python
print('Price Range:')
print('Maximum', round(data.describe().loc['max','Price range'],2))
print('Minimum', round(data.describe().loc['min','Price range'],2))
```

```python
print('Aggregate rating:')
print('Maximum', round(data.describe().loc['max','Aggregate rating'],2))
print('Minimum', round(data.describe().loc['min','Aggregate rating'],2))
print("Total Number of Votes in the rating:", int(data.describe().loc['count','Votes']))
```

```python
# Let us find out how many restaurants are given based on the countries:

data.groupby(['Country']).size().sort_values(ascending=False)
```

```python
# Another Method:
data['Country'].value_counts()
```

```python
# PLOTTING PIE CHART BASED ON THE ABOVE OBSERVATION BETWEEN TOP 4 COUNTRIES AND BOTTOM 5 COUNTRIES 
# Top 4
plt.pie(x = data['Country'].value_counts().values[:4], labels=data['Country'].value_counts().index[:4], autopct="%1.2f%%")
plt.plot
```

```python
# Bottom Five
plt.pie(x = data['Country'].value_counts().values[-5:], labels=data['Country'].value_counts().index[-5:], autopct="%1.2f%%")
plt.plot
```

```python
# Working on the Ratings Now:
ratings = data.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings 
```

```python
# Renaming the rating count from 0 to Rating Count and Saving it to the ratings itself
ratings = ratings.reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## OBSERVATION: 
Maximum amount of people have not rated the restaurants

```python
# Now let us group the data based on colour and text and plot the graph
total_ratings = ratings.groupby(['Rating color','Rating text']).sum().reset_index()
```

```python
total_ratings
```

```python
# plotting the graph based on the count 
sns.barplot(x=total_ratings['Rating text'], y=total_ratings['Rating Count'], palette=['Blue', 'Green', 'Orange', 'Red','Black','Yellow'])
```

## OBSERVATION:
Here people have rated as Average most and the poor least on the following scale:


0 - Not Rated

Below 2.4 - Poor

2.4 - 3.5 - Average

3.5 - 3.9 - Good

4 - 4.4 - Very Good

Above 4.4 - Excellent
